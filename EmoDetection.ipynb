```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
import numpy as np
import cv2
import base64

# Function to capture an image from the webcam
def capture_image():
    js = Javascript('''
        async function captureImage() {
            const div = document.createElement('div');
            document.body.appendChild(div);
            const video = document.createElement('video');
            video.style.display = 'block';
            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            div.appendChild(video);
            video.srcObject = stream;
            await video.play();

            // Resize video for better view in Colab
            video.width = 640;
            video.height = 480;

            // Capture image
            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            const context = canvas.getContext('2d');
            context.drawImage(video, 0, 0, canvas.width, canvas.height);
            stream.getTracks().forEach(track => track.stop());
            div.remove();

            return canvas.toDataURL('image/jpeg', 0.8);
        }
        captureImage();
    ''')
    display(js)
    data = eval_js("captureImage()")
    binary = base64.b64decode(data.split(',')[1])
    img = np.frombuffer(binary, dtype=np.uint8)
    return cv2.imdecode(img, cv2.IMREAD_COLOR)

# Capture image from webcam
frame = capture_image()
```

```python
# Install the FER library
!pip install fer
from google.colab.patches import cv2_imshow  # Import this to display images in Colab


from fer import FER

# Create a FER detector instance
detector = FER()

# Analyze the captured frame
result = detector.detect_emotions(frame)

# Get the dominant emotion
if result:
    detected_emotion = max(result[0]['emotions'], key=result[0]['emotions'].get)
    print("Detected emotion:", detected_emotion)

    # Display the image with the detected emotion
    cv2.putText(frame, f"Emotion: {detected_emotion}", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2_imshow(frame)  # Show the image with detected emotion
else:
    print("No emotion detected.")
```
